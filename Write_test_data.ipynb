```python
import numpy as np
import pandas as pd

import os, glob

from collections import OrderedDict

import matplotlib
import matplotlib.pyplot as plt

%matplotlib nbagg
```

```python
plt.figure()
ax1 = plt.subplot(111)
spectra = OrderedDict()

cens = np.linspace(48, 52, 20)
peaks = 1 + np.random.rand(len(cens))

x = np.arange(200)
for i, (cen, peak) in enumerate(zip(cens, peaks)): 
    y = peak*np.exp(-(x-cen)**2 / 20) + np.random.rand(len(x))  + 2*peak*np.exp(-(x-100)**2 / 20) 
    spectra.update({"{:0>4d}".format(i): pd.Series(y, index=x, name="{:0>4d}".format(i))})

for name, spectrum in spectra.items():
    spectrum.plot()

plt.legend()


```

```python
for name, spectrum in spectra.items():
    x = spectrum.index
    y = spectrum.values
    M = np.vstack((x,y,y)).transpose()
    np.savetxt('test_data/{}.txt'.format(name), M)
    
```
